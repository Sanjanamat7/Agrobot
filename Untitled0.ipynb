```python
from ultralytics import YOLO

# Load a YOLOv8 pre-trained model
model = YOLO("yolov8n.pt")   # 'n' = nano (fastest), can also use 's', 'm', 'l'

# Train the model
model.train(
    data=dataset.location + "/data.yaml",  # path to dataset
    epochs=50,        # number of training epochs
    imgsz=640,        # image size
    batch=16          # batch size
)
```

```python
# Validate model on test set
metrics = model.val()
print(metrics)
```
